```python
import pandas as pd
import matplotlib.pyplot as plt
from utils import read_data
from sklearn.linear_model import LinearRegression
from prediction import regression_task
from visulization import *
from tsai.all import *
import ray
from ray import tune
from ray.tune.schedulers import ASHAScheduler
import warnings
warnings.filterwarnings('ignore')
```

```python
exchanges = read_data(folder_path='../Exp Data/v2/', name='exchange', binary=False)
exchanges = exchanges[exchanges.index >= '2015-01-01']
cryptos = ['USDBTC', 'USDDOGE', 'USDETH', 'USDLTC', 'USDXRP']
convens = ['USDEUR', 'USDJPY', 'USDGBP', 'USDAUD', 'USDCAD', 'USDCHF', 'USDCNY', 'USDSEK', 'USDNZD']
exchanges = exchanges[cryptos+convens]
```

```python
FOCAL = 'BTC'
FEATURE_NUM = 14
LAG = 7
HORIZION = 1
SPLITS=30
LOOK_BACK=3*365+SPLITS
JUMP=30
```

```python
# hyperparameter tunning
search_space = {
    "hidden_size": tune.grid_search([20, 50, 100, 200]),
    "n_layers": tune.grid_search([1, 2 ,3 ,4]),
    "rnn_dropout": tune.grid_search([0, 0.1, 0.2, 0.3])
}

def tunning_job(config):
    clf = LSTM(LAG, HORIZION, hidden_size=config['hidden_size'], n_layers=config['n_layers'],
               bias=True, rnn_dropout=config['rnn_dropout'], bidirectional=True)
    loss_all, loss_part  = regression_task(exchanges, 'BTC', cryptos, model, HORIZION, SPLITS, LOOK_BACK, LAG, jump=JUMP)
    tune.report(mean_mae_withc=np.mean(loss_all), mean_mae_withoutc=np.mean(loss_part))

asha_scheduler = ASHAScheduler(time_attr="training_iteration", metric="mean_mae_withc", mode="max")

analysis = tune.run(tunning_job, config=search_space, name='BiLSTMs', scheduler=asha_scheduler,
                        stop={'training_iteration': 100}, max_failures=10, verbose=1, resume="AUTO")
```

```python
model = LSTM(LAG, HORIZION, hidden_size=100, n_layers=2, bias=True, rnn_dropout=0.2, bidirectional=True)
loss_all, loss_part = regression_task(exchanges, 'BTC', cryptos, model, HORIZION, SPLITS, LOOK_BACK, LAG, jump=JUMP)
```

```python
# full
mae_loss = []
for i in tqdm(range(LOOK_BACK, whole_dataset.shape[0]-SPLITS, JUMP)):
    sup_x, sup_y = [], []
    X, y = seq2seq_X_y(whole_dataset, FEATURE_NUM, i, LAG, LOOK_BACK, HORIZION)
    # to TSTensor
    X, y = TSTensor(X), TSTensor(y)
    # create the model
    model = LSTM(LAG, HORIZION, hidden_size=100, n_layers=2, bias=True, rnn_dropout=0.2, bidirectional=True)
    # learner
    batch_tfms = TSNormalize()
    splits = TSSplitter(SPLITS)(y)
    dls = get_ts_dls(X, y, splits=splits, batch_tfms=batch_tfms)
    learner = ts_learner(dls, model, metrics=mae, verbose=False)
    # fit
    learner.fit_one_cycle(20, 1e-2)
    # append results
    mae_loss.append(learner.final_record[-1])
    # clear output
    clear_output(wait=True)
```
