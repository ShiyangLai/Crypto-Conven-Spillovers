## __Logistic Regression__

```python
search_space = {
    "penalty": tune.grid_search(['l1', 'l2']),
    "C": tune.grid_search([0.1, 1, 10, 100, 1000]),
    "solver": tune.grid_search(['saga'])
}

def tunning_job(config):
    clf = LogisticRegression(random_state=36, penalty=config['penalty'], solver=config['solver'], C=config['C'])
    predicted_all, predicted_part, real = binary_task(returns, 'BTC', cryptos, clf, lookback=3*365, jump=10, step_ahead=90)
    _all1, _part1 = dynamic_evaluation(real, predicted_all, predicted_part, metrics.accuracy_score)
    _all2, _part2 = dynamic_evaluation(real, predicted_all, predicted_part, false_positive_rate)
    _all3, _part3 = dynamic_evaluation(real, predicted_all, predicted_part, false_negative_rate)
    _all4, _part4 = dynamic_evaluation(real, predicted_all, predicted_part, metrics.precision_score)
    _all5, _part5 = dynamic_evaluation(real, predicted_all, predicted_part, metrics.recall_score)
    tune.report(mean_accuracy_withc=np.mean(_all1), mean_accuracy_withoutc=np.mean(_part1), 
                mean_fpr_withc=np.mean(_all2), mean_fpr_withoutc=np.mean(_part2),
                mean_fnr_withc=np.mean(_all3), mean_fnr_withoutc=np.mean(_part3),
                mean_precision_withc=np.mean(_all4), mean_precision_withoutc=np.mean(_part4),
                mean_recall_withc=np.mean(_all5), mean_recall_withoutc=np.mean(_part5))

asha_scheduler = ASHAScheduler(time_attr="training_iteration", metric="mean_accuracy_withc", mode="max")

analysis = tune.run(tunning_job, config=search_space, name='Logistic Regression', scheduler=asha_scheduler,
                        stop={'training_iteration': 100}, max_failures=10, verbose=1, resume="AUTO")
```

## __Support Vector Machine__

```python
search_space = {
    "kernel": tune.grid_search(['linear', 'rbf', 'poly']),
    "gamma": tune.grid_search([0.1, 1, 10, 100]),
    "C": tune.grid_search([0.1, 1, 10, 100]),
    "degree": tune.grid_search([2, 3, 4])
}

def tunning_job(config):
    clf = SVC(random_state=36, kernel=config['kernel'], gamma=config['gamma'], C=config['C'], degree=config['degree'])
    predicted_all, predicted_part, real = binary_task(returns, 'BTC', cryptos, clf, lookback=3*365, jump=10, step_ahead=90)
    _all1, _part1 = dynamic_evaluation(real, predicted_all, predicted_part, metrics.accuracy_score)
    _all2, _part2 = dynamic_evaluation(real, predicted_all, predicted_part, false_positive_rate)
    _all3, _part3 = dynamic_evaluation(real, predicted_all, predicted_part, false_negative_rate)
    _all4, _part4 = dynamic_evaluation(real, predicted_all, predicted_part, metrics.precision_score)
    _all5, _part5 = dynamic_evaluation(real, predicted_all, predicted_part, metrics.recall_score)
    tune.report(mean_accuracy_withc=np.mean(_all1), mean_accuracy_withoutc=np.mean(_part1), 
                mean_fpr_withc=np.mean(_all2), mean_fpr_withoutc=np.mean(_part2),
                mean_fnr_withc=np.mean(_all3), mean_fnr_withoutc=np.mean(_part3),
                mean_precision_withc=np.mean(_all4), mean_precision_withoutc=np.mean(_part4),
                mean_recall_withc=np.mean(_all5), mean_recall_withoutc=np.mean(_part5))
    

asha_scheduler = ASHAScheduler(time_attr="training_iteration", metric="mean_accuracy_withc", mode="max")

analysis = tune.run(tunning_job, config=search_space, name='Support Vector Machine', metric="mean_accuracy_withc", mode="max",
                        stop={'training_iteration': 200}, max_failures=2, verbose=1, resume="AUTO")
```

## __Random Forest__

```python
search_space = {
    "n_estimators": tune.grid_search([10, 100, 500]),
    "criterion": tune.grid_search(['gini', 'entropy']),
    "max_depth": tune.grid_search([10, 50, 100, None]),
    "min_samples_leaf": tune.grid_search([1, 10, 100]),
}

def tunning_job(config):
    clf = RandomForestClassifier(random_state=36, n_estimators=config['n_estimators'], criterion=config['criterion'], max_depth=config['max_depth'], min_samples_leaf=config['min_samples_leaf'])
    predicted_all, predicted_part, real = binary_task(returns, 'BTC', cryptos, clf, lookback=3*365, jump=10, step_ahead=90)
    _all1, _part1 = dynamic_evaluation(real, predicted_all, predicted_part, metrics.accuracy_score)
    _all2, _part2 = dynamic_evaluation(real, predicted_all, predicted_part, false_positive_rate)
    _all3, _part3 = dynamic_evaluation(real, predicted_all, predicted_part, false_negative_rate)
    _all4, _part4 = dynamic_evaluation(real, predicted_all, predicted_part, metrics.precision_score)
    _all5, _part5 = dynamic_evaluation(real, predicted_all, predicted_part, metrics.recall_score)
    tune.report(mean_accuracy_withc=np.mean(_all1), mean_accuracy_withoutc=np.mean(_part1), 
                mean_fpr_withc=np.mean(_all2), mean_fpr_withoutc=np.mean(_part2),
                mean_fnr_withc=np.mean(_all3), mean_fnr_withoutc=np.mean(_part3),
                mean_precision_withc=np.mean(_all4), mean_precision_withoutc=np.mean(_part4),
                mean_recall_withc=np.mean(_all5), mean_recall_withoutc=np.mean(_part5))

asha_scheduler = ASHAScheduler(time_attr="training_iteration", metric="mean_accuracy_withc", mode="max")

analysis = tune.run(tunning_job, config=search_space, name='Random Forest', scheduler=asha_scheduler,
                        stop={'training_iteration': 200}, max_failures=5, verbose=1, resume="AUTO")
```

## __Gate Recurrent Unit__

```python
search_space = {
    "lr": tune.loguniform(0.001, 0.1),
    "gru_1": tune.randint(2, 128),
    "drop_1": tune.uniform(0, 0.5),
}

def tunning_job(config):
    optimizer = optimizers.Adam(learning_rate=config['lr'])
    clf = grus(1, 1, activation='sigmoid', optimizer=optimizer, gru1=config['gru_1'], drop1=config['drop_1'])
    predicted_all, predicted_part, real = binary_task(returns, 'BTC', cryptos, clf, lookback=3*365, jump=10, step_ahead=90, optimizer=optimizer, epochs=20, verbose=0)
    _all1, _part1 = dynamic_evaluation(real, predicted_all, predicted_part, metrics.accuracy_score)
    _all2, _part2 = dynamic_evaluation(real, predicted_all, predicted_part, false_positive_rate)
    _all3, _part3 = dynamic_evaluation(real, predicted_all, predicted_part, false_negative_rate)
    _all4, _part4 = dynamic_evaluation(real, predicted_all, predicted_part, metrics.precision_score)
    _all5, _part5 = dynamic_evaluation(real, predicted_all, predicted_part, metrics.recall_score)
    tune.report(mean_accuracy_withc=np.mean(_all1), mean_accuracy_withoutc=np.mean(_part1), 
                mean_fpr_withc=np.mean(_all2), mean_fpr_withoutc=np.mean(_part2),
                mean_fnr_withc=np.mean(_all3), mean_fnr_withoutc=np.mean(_part3),
                mean_precision_withc=np.mean(_all4), mean_precision_withoutc=np.mean(_part4),
                mean_recall_withc=np.mean(_all5), mean_recall_withoutc=np.mean(_part5))

asha_scheduler = ASHAScheduler(time_attr="training_iteration", metric="mean_accuracy_withc", mode="max")

analysis = tune.run(tunning_job, config=search_space, name='GRUs', scheduler=asha_scheduler, num_samples=30,
                        stop={'training_iteration': 200}, max_failures=5, verbose=1, resume="AUTO")
```
